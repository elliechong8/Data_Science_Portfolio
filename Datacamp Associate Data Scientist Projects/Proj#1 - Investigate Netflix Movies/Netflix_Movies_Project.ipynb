### Exploratory Data Analysis

Question 1: What was the most frequent movie duration in the 1990s?

```python
# Filter for movies released in the 1990s
movies_90s = netflix_df[
    (netflix_df["type"] == "Movie") &
    (netflix_df["release_year"] >= 1990) &
    (netflix_df["release_year"] <= 1999)]

movies_90s
```

```python
# Find the most frequent movie duration in the 1990s
duration = int(movies_90s["duration"].mode())
print(f"The most frequent movie duration in the 1990s is: {duration}.")
```

```python
# Viewing the most frequent movie duration in histogram
plt.hist(movies_90s["duration"].mode())
plt.show()
```

Task 2: Count the number of short action movies released in the 1990s

A movie is considered short if it is less than 90 minutes. 

```python
# Filter for short action movies released in the 1990s and count the number of short action movies
short_movie_count = movies_90s[
    (movies_90s["genre"] == "Action") &
    (movies_90s["duration"] <= 90 )
].shape[0]

print(f"The number of short action movies released in the 1990s is: {short_movie_count}.")
```
